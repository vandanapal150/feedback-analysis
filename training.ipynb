```python
import pandas as pd
```

```python
list_feedback=[]
list_target=[]
with open('data/en-training.txt') as dev:
    for i in dev:
        feedback=i.strip().lower().split()
        if len(feedback) >= 1:
            target=feedback[-1]
            del(feedback[0])
            del(feedback[-1])
            list_feedback.append(feedback)
            list_target.append(target)
```

```python
list_feedback
```

```python
list_target
```

```python
stopwords=[]
with open('data/stopwords.txt') as stops:
    for i in stops:
        stopwords.append(i.strip())
stopwords
```

```python
fdbk=[]
for i in list_feedback:
    refined_fd=[]
    for j in i:
        if j not in stopwords:
            refined_fd.append(j)
    fdbk.append(refined_fd)
fdbk
```

```python
len(list_target),len(list_feedback)
```

```python
from collections import defaultdict
d=defaultdict(int)
refnd_fdb=[]
for i in fdbk:
    for j in i:
        d[j]+=1
for i in fdbk:
    temp_fd=[]
    for j in i:
        if d[j]>1:
            temp_fd.append(j)
    refnd_fdb.append(temp_fd)
refnd_fdb
```

```python
count=0
for i in refnd_fdb:
    if len(i)==0:
        refnd_fdb.remove(i)
        list_target.pop(count)
    count+=1
refnd_fdb
```

```python
len(list_target),len(refnd_fdb)
```

```python
import gensim
d2=gensim.corpora.Dictionary(refnd_fdb)
bow_list=[]
for i in refnd_fdb:
    bow_list.append(d2.doc2bow(i))
bow_list
```

```python
cnt=len(d2)
cnt
dense_bow=gensim.matutils.corpus2dense(bow_list,num_terms=cnt).transpose()
dense_bow
```

```python
list_target
refnd_trg=[]
for i in list_target:
    if i=='comment':
        refnd_trg.append(0)
    elif i=='complaint':
        refnd_trg.append(1)
    elif i=='meaningless':
        refnd_trg.append(2)
    elif i=='request':
        refnd_trg.append(3)
    elif i=='bug':
        refnd_trg.append(4)
    else:
        refnd_trg.append(5)
refnd_trg        
```

```python
len(refnd_trg)
```

```python
tfidf=gensim.models.TfidfModel(bow_list)
tfidf
```

```python
tfidf_list=tfidf[bow_list]
tfidf
```

```python
dense_tfidf=gensim.matutils.corpus2dense(tfidf_list,num_terms=count).transpose()
dense_tfidf
```

```python
import sklearn
from sklearn.model_selection import KFold
kf=KFold(n_splits=10,shuffle=True)
dataset=dense_tfidf
sd=list(kf.split(dataset))
sd
```

```python
import numpy as np
from sklearn import svm
#from sklearn.naive_bayes import GaussianNB
#from sklearn.naive_bayes import BernoulliNB
#from sklearn.naive_bayes import MultinomialNB
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
#from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
acc_list=[]
f1_scores = []
precision_scores=[]
recall_scores=[]
for test,train in sd:
    train_data=[]
    train_label=[]
    test_data=[]
    test_label=[]
    for i in train:
        train_data.append(dataset[i])
        train_label.append(refnd_trg[i])
    for j in test:
        test_data.append(dataset[j])
        test_label.append(refnd_trg[j])
    classifier=svm.SVC()
    #classifier=GaussianNB()
    #classifier=BernoulliNB()
    #classifier=MultinomialNB()
    #classifier=KNeighborsClassifier(n_neighbors=10)
    
    classifier = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
    classifier = classifier.fit(train_data, train_label)
    predict=classifier.predict(test_data)
#     print(type(test_label),type())
    incorrect=(predict!=test_label).sum()
    score = f1_score(test_label, predict, average='weighted')
    score1 = precision_score(test_label, predict, average='weighted')
    score2 = recall_score(test_label, predict, average='weighted')
#     print(score)
    f1_scores.append(score)
    precision_scores.append(score1)
    recall_scores.append(score2)
    acc=(len(test_label)-incorrect)/len(test_label)
    acc_list.append(acc)
```
