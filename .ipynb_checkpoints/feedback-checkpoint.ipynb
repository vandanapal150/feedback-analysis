```python
import pandas as pd
```

```python
list_feedback=[]
list_target=[]
with open('data/en-development.txt') as dev:
    for i in dev:
        feedback=i.strip().lower().split()
        if len(feedback) >= 1:
            target=feedback[-1]
            del(feedback[0])
            del(feedback[-1])
            list_feedback.append(feedback)
            list_target.append(target)
```

```python
list_feedback
```

```python
list_target
```

```python
stopwords=[]
with open('data/stopwords.txt') as stops:
    for i in stops:
        stopwords.append(i.strip())
stopwords
```

```python
fdbk=[]
refined_fd=[]
for i in list_feedback:
    for j in i:
        if j not in stopwords:
            refined_fd.append(i)
            fdbk.append(refined_fd)
fdbk
```

```python
from collections import defaultdict
d=defaultdict(int)
for i in list_feedback:
    for j in i:
        d['j']=d['j']+1
d
```
